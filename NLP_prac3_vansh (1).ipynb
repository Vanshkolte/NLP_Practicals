```python
# Function to segment text into words
def segment_words(text):
    return re.findall(r'\b\w+\b', text)

words = segment_words(headline)

print("Words:")
print(words[:20])
```

```python
# Function to segment text into sentences
def segment_sentences(text):
    sentence_pattern = r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s'
    return re.split(sentence_pattern, text)

sentences = segment_sentences(headline)

print("\nSentences:")
for i, sentence in enumerate(sentences, 1):
    print(f"{i}. {sentence}")
```

```python
# 4. POS tagging (simple rule-based with regex)
def pos_tagging(words):
    pos_tags = []
    for word in words:
        if re.fullmatch(r'[A-Z][a-z]*', word):  # Starts with capital: Proper Noun
            pos_tags.append((word, 'NNP'))
        elif re.fullmatch(r'\d+', word):  # Numbers: Numeric
            pos_tags.append((word, 'NUM'))
        elif re.fullmatch(r'\w+ing\b', word):  # Ends with 'ing': Gerund/Verb
            pos_tags.append((word, 'VBG'))
        elif re.fullmatch(r'\w+ed\b', word):  # Ends with 'ed': Past Tense Verb
            pos_tags.append((word, 'VBD'))
        elif re.fullmatch(r'\w+s\b', word):  # Ends with 's': Plural Noun or Verb
            pos_tags.append((word, 'NNS'))
        else:  # Default to common noun
            pos_tags.append((word, 'NN'))
    return pos_tags

pos_tags = pos_tagging(words)
print("\nPOS Tags:", pos_tags)
```

```python
def stem_word(word):
    if word.endswith('ing') or word.endswith('ed') or word.endswith('es') or word.endswith('s'):
        return re.sub(r'(ing|ed|es|s)$', '', word)
    return word

stemmed_words = [stem_word(word) for word in words]
print("\nStemmed Words:", stemmed_words)
```

```python
def get_unique_words(words):
    return set(words)

unique_words = get_unique_words(words)
print("\nUnique Words:", unique_words)
```

```python
import re

# Common stop words
stop_words = [
    "a", "an", "and", "are", "as", "at", "be", "by", "for", "from", "has",
    "he", "in", "is", "it", "its", "of", "on", "that", "the", "to", "was",
    "were", "will", "with", "she", "her", "his", "they", "you", "your", "but",
    "or", "if", "we", "this", "there"
]

# Convert the stop words to a regex pattern
stop_word_pattern = r'\b(' + '|'.join(map(re.escape, stop_words)) + r')\b'

# Function to find stop words in text
def find_stop_words(text):
    return re.findall(stop_word_pattern, text, flags=re.IGNORECASE)

# Identify stop words in the text
stop_words_found = find_stop_words(headline)

# Print the results
print("Stop words found:")
print(stop_words_found)
print("Number of stop_words found : ", str(len(stop_words)))
```
