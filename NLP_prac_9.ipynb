```python
model_name = "Helsinki-NLP/opus-mt-en-hi" # model is use to translate english to hindi
```

```python
from transformers import MarianMTModel, MarianTokenizer # Import MarianTokenizer

# load the model
model = MarianMTModel.from_pretrained(model_name)

tokenizer = MarianTokenizer.from_pretrained(model_name) # Now MarianTokenizer is defined
```

```python
text = "you are the one"
```

```python
token = tokenizer(text, return_tensors="pt") # Tokenize the input text
```

```python
translated_text =tokenizer.decode(model.generate(**token)[0], skip_special_tokens=True) # Generate the translation
```

```python
print(f"Original Text: {text}")
print(f"Translated Text: {translated_text}")
```
