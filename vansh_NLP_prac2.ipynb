# Rule based POS Tagging

```python
import re

# Define basic rules
rules = {}

def tokenize(text):
  return text.split()

def pos_tag(sentence):
  words = tokenize(sentence)
  tagged = []

  for word in words:
    tagged_word = (word, 'NN')  # Initialize with default tag 'NN'
    for tag, rule in rules.items():
      if re.match(rule, word):
        tagged_word = (word, tag)  # Update tag if rule matches
        break

    tagged.append(tagged_word)  # Append the tagged word

  return tagged

# Get user input
sentence = input("Enter a sentence for POS tagging: ")

# Get POS tags for the sentence
tags = pos_tag(sentence)

# Output the POS tags
print("POS Tagging Result:")
for word, tag in tags:  # Use 'word' instead of 'words'
  print(f"{word}: {tag}")
```

```python
import re

# Define basic rules
# NN=> Noun    VB=>Verb    JJ=>Adjective     RB=> Adverd     DT=> Determiner
rules = {
    'NN' : r'\b[a-zA-Z]+(?:ness|ment|ity|s)\b',
    'PRP' : r'\b(?:I|you|he|she|it|we|they)\b',
    'JJ' : r'\b[a-zA-Z]+(?:ly|ed|ing)\b',
}

def tokenize(text):
  return text.split()

def pos_tag(sentence):
  words = tokenize(sentence)
  tagged = []

  for word in words:
    tagged_word = (word, 'NN')  # Initialize with default tag 'NN'
    for tag, rule in rules.items():
      if re.match(rule, word):
        tagged_word = (word, tag)  # Update tag if rule matches
        break

    tagged.append(tagged_word)  # Append the tagged word

  return tagged

# Get user input
sentence = input("Enter a sentence for POS tagging: ")

# Get POS tags for the sentence
tags = pos_tag(sentence)

# Output the POS tags
print("POS Tagging Result:")
for word, tag in tags:  # Use 'word' instead of 'words'
  print(f"{word}: {tag}")
```

```python
def preprocess(sentence):
  #Tokenize the sentence into words
  words = word_tokenize(sentence)
  #Remove punctuation and convert to lowercase
  words = [word for word in words if word not in string.punctuation]
  #Remove stopwords
  words = [word for word in words if word.lower() not in stop_words]
  #Perform stemming
  words = [stemmer.stem(word) for word in words]
  #perform POS tagging
  tagged_words = pos_tag(words)
  return tagged_words
```

```python
for sentence in sentences:
  tagged_words = preprocess(sentence)[:20]
print(tagged_words, "/n")
```
